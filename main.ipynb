```python
# Let's get started. First, install the Python Packages that we need such as `pandas`, `numpy`, `matplotlib`, and `seaborn` or anything else.
# And then import it into our codebase:

import pandas as pd
import numpy as np
import matplotlib    # For checking Matplotlib version
import matplotlib.pyplot as plt
import seaborn as sns

# Let's check if all the packages are installed correctly:
print(f'Pandas version: {pd.__version__}')
print(f'Numpy version: {np.__version__}')
print(f'Matplotlib version: {matplotlib.__version__}')
print(f'Seaborn version: {sns.__version__}')
```

```python
# Now that all the necessary packages are installed correctly, let's learn how to read data from your spreadsheets into our code.

# First, define the URL of the spreadsheet. Make sure that the spreadsheet is accessible to the public:
sheet_url = 'https://docs.google.com/spreadsheets/d/1GdgYeJCOzB_A-KBcwgvSlV2lDjYLFcGmKAyWBNDVJjg/edit?gid=1903509374#gid=1903509374'

# Next, change the URL to export the spreadsheet as a CSV file:
sheet_url_trf = sheet_url.replace('/edit?gid=', '/export?format=csv&gid=')

# Now, let's read the spreadsheet and display the first few rows:
print(sheet_url_trf)
df = pd.read_csv(sheet_url_trf)
df.head()
```

```python
# Let's check the information of the dataframe:
df.info()
```

```python
# If we have successfully defined the spreadsheet and read the dataframe, let's proceed to clean the data to improve its readability.

# First, let's create a copy of the dataframe:
df_cleaned = df.copy()

# Next, convert the `Date` column to datetime format:
df_cleaned['Date'] = pd.to_datetime(df_cleaned['Date'])

# Now, let's calculate the Revenue by multiplying the Unit price and Quantity data:
df_cleaned['Revenue'] = df_cleaned['Unit price'] * df_cleaned['Quantity']

# Finally, let's inspect the cleaned dataframe:
df_cleaned
```

```python
# Let's check the information of the cleaned dataframe:
df_cleaned.info()
```

```python
# We can check the value counts of the unique values in the dataframe, for example we can check the unique value counts of `Gender` column:
df_cleaned['Gender'].value_counts()
```

```python
# If we want to check the unique value counts of all the columns in the dataframe, we can use a loop and fetch the colums using array:
for col_name in ['Payment', 'Branch', 'City', 'Customer type', 'Gender', 'Product line']:
    print(df_cleaned[col_name].value_counts(),'\n')
```

```python
# In case there is so many typos in the data from a column, We can also fix them. For example, we can correct typos in the `Gender` column.

# First, let's define a dictionary that maps common typos to the correct values:
dict_typo = {'male' : 'Male',
            'female' : 'Female',
            'fml' : 'Female'}

# Next, we use the replace method to apply these corrections to the DataFrame:
df_replaced = df.cleaned.replace(dict_typo)

# Finally, let's take a look at the first few rows of the corrected DataFrame:
df_replaced.head()
```

```python
# Now, let's check if there are any duplicated values in the `Invoice ID` column:
df_replaced['Invoice ID'].duplicated().sum()

# In this case, there are no duplicated values:
df_replaced.head()
```

```python
# To analyze daily revenue and quantity from the cleaned data, we can use the groupby method to aggregate the data.

# Group by 'Date' and aggregate 'Revenue' and 'Quantity' columns:
agg_daily = df_replaced.groupby('Date', as_index=False)[['Revenue', 'Quantity']].sum()

# Display the first few rows of the aggregated dataframe:
agg_daily.head()
```

```python
# Or if you want to visualize the Daily Revenue Trends data, you can use the Seaborn and Matplot

# Set the figure size for plots:
plt.rcParams['figure.figsize'] = (20, 5)

# Plot a line graph using Seaborn:
sns.lineplot(data = agg_daily, x = 'Date', y = 'Revenue')
```

```python
# Here's another example of analyzing payments from the cleaned data

# Group by 'Payment' method and aggregate 'Revenue' and count unique 'Invoice ID':
agg_payment = df_replaced.groupby('Payment', as_index=False).agg({'Revenue' : 'sum', 'Invoice ID' : 'nunique'})

# Display the aggregated payment data:
agg_payment
```

```python
# Visualize the data using Seaborn and Matplot

# Set the figure size for plots:
plt.rcParams['figure.figsize'] = (10, 5)

# Plot a bar graph using Seaborn:
sns.barplot(x = 'Payment', y = 'Revenue', data = agg_payment.sort_values('Revenue', ascending=False))
```

```python
# If you want to visualize the data with a different graph styles for example a pie chart, you can use Matplotlib Pie Chart.

# Plot a pie chart using Matplotlib:
plt.pie(agg_payment['Revenue'], labels = agg_payment['Payment'], autopct='%1.1f%%')
```

```python
# Let's try another example.

# Group by 'Gender' and 'Product line', then count unique 'Invoice ID'
agg_gender = df_replaced.groupby(['Gender', 'Product line'], as_index=False)['Invoice ID'].nunique()

# Display the first few rows of the aggregated dataframe:
agg_gender.head()
```

```python
# Visualize the data using Seaborn and Matplot

# Set the figure size for plots:
plt.rcParams['figure.figsize'] = (20, 5)

# Plot a bar graph using Seaborn:
sns.barplot(x = 'Product line', y = 'Invoice ID', data = agg_gender, hue = 'Gender')
```
